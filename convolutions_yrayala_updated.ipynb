In this notebook, we'll walk through how convolutional filters can pick out different aspects of an image

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from skimage import data
from skimage.color import rgb2gray
from skimage.transform import resize
```

## 0. Original Input Image <a name="orig"></a>

```python
im = rgb2gray(data.coffee())
im = resize(im, (64, 64))
print(im.shape)

plt.axis("off")
plt.imshow(im, cmap="gray")
```

## Horizontal Edge Filter <a name="hor"></a>

```python
filter1 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

new_image = np.zeros(im.shape)

im_pad = np.pad(im, 1, "constant")

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        try:
            new_image[i, j] = (
                im_pad[i - 1, j - 1] * filter1[0, 0]
                + im_pad[i - 1, j] * filter1[0, 1]
                + im_pad[i - 1, j + 1] * filter1[0, 2]
                + im_pad[i, j - 1] * filter1[1, 0]
                + im_pad[i, j] * filter1[1, 1]
                + im_pad[i, j + 1] * filter1[1, 2]
                + im_pad[i + 1, j - 1] * filter1[2, 0]
                + im_pad[i + 1, j] * filter1[2, 1]
                + im_pad[i + 1, j + 1] * filter1[2, 2]
            )
        except:
            pass

plt.axis("off")
plt.imshow(new_image, cmap="Greys")
```

## Vertical Edge Filter <a name="ver"></a>

```python
filter2 = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

new_image = np.zeros(im.shape)

im_pad = np.pad(im, 1, "constant")

for i in range(im.shape[0]):
    for j in range(im.shape[1]):
        try:
            new_image[i, j] = (
                im_pad[i - 1, j - 1] * filter2[0, 0]
                + im_pad[i - 1, j] * filter2[0, 1]
                + im_pad[i - 1, j + 1] * filter2[0, 2]
                + im_pad[i, j - 1] * filter2[1, 0]
                + im_pad[i, j] * filter2[1, 1]
                + im_pad[i, j + 1] * filter2[1, 2]
                + im_pad[i + 1, j - 1] * filter2[2, 0]
                + im_pad[i + 1, j] * filter2[2, 1]
                + im_pad[i + 1, j + 1] * filter2[2, 2]
            )
        except:
            pass

plt.axis("off")
plt.imshow(new_image, cmap="Greys")
```

## Horizontal Edge Filter with Stride 2 <a name="hor2"></a>

```python
filter1 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

stride = 2

new_image = np.zeros((int(im.shape[0] / stride), int(im.shape[1] / stride)))

im_pad = np.pad(im, 1, "constant")

for i in range(0, im.shape[0], stride):
    for j in range(0, im.shape[1], stride):
        try:
            new_image[int(i / stride), int(j / stride)] = (
                im_pad[i - 1, j - 1] * filter1[0, 0]
                + im_pad[i - 1, j] * filter1[0, 1]
                + im_pad[i - 1, j + 1] * filter1[0, 2]
                + im_pad[i, j - 1] * filter1[1, 0]
                + im_pad[i, j] * filter1[1, 1]
                + im_pad[i, j + 1] * filter1[1, 2]
                + im_pad[i + 1, j - 1] * filter1[2, 0]
                + im_pad[i + 1, j] * filter1[2, 1]
                + im_pad[i + 1, j + 1] * filter1[2, 2]
            )
        except:
            pass

plt.axis("off")
plt.imshow(new_image, cmap="Greys")
```

## Vertical Edge Filter with Stride 2 <a name="ver2"></a>

```python
filter2 = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

stride = 2

new_image = np.zeros((int(im.shape[0] / stride), int(im.shape[1] / stride)))

im_pad = np.pad(im, 1, "constant")

for i in range(0, im.shape[0], stride):
    for j in range(0, im.shape[1], stride):
        try:
            new_image[int(i / stride), int(j / stride)] = (
                im_pad[i - 1, j - 1] * filter2[0, 0]
                + im_pad[i - 1, j] * filter2[0, 1]
                + im_pad[i - 1, j + 1] * filter2[0, 2]
                + im_pad[i, j - 1] * filter2[1, 0]
                + im_pad[i, j] * filter2[1, 1]
                + im_pad[i, j + 1] * filter2[1, 2]
                + im_pad[i + 1, j - 1] * filter2[2, 0]
                + im_pad[i + 1, j] * filter2[2, 1]
                + im_pad[i + 1, j + 1] * filter2[2, 2]
            )
        except:
            pass

plt.axis("off")
plt.imshow(new_image, cmap="Greys")
```

## Horizontal Edge Filter with Stride 3 (Custom) <a name="hor3"></a>

```python
filter1 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])

stride = 3

new_image_stride3 = np.zeros((int(im.shape[0] / stride) + 1, int(im.shape[1] / stride) + 1))

im_pad = np.pad(im, 1, "constant")

for i in range(0, im.shape[0], stride):
    for j in range(0, im.shape[1], stride):
        try:
            new_image_stride3[int(i / stride), int(j / stride)] = (
                im_pad[i - 1, j - 1] * filter1[0, 0]
                + im_pad[i - 1, j] * filter1[0, 1]
                + im_pad[i - 1, j + 1] * filter1[0, 2]
                + im_pad[i, j - 1] * filter1[1, 0]
                + im_pad[i, j] * filter1[1, 1]
                + im_pad[i, j + 1] * filter1[1, 2]
                + im_pad[i + 1, j - 1] * filter1[2, 0]
                + im_pad[i + 1, j] * filter1[2, 1]
                + im_pad[i + 1, j + 1] * filter1[2, 2]
            )
        except:
            pass

plt.axis("off")
plt.imshow(new_image_stride3, cmap="Greys")
print(f"Output shape with stride=3: {new_image_stride3.shape}")
```

## Vertical Edge Filter with Stride 3 (Custom) <a name="ver3"></a>

```python
filter2 = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])

stride = 3

new_image_stride3_vert = np.zeros((int(im.shape[0] / stride) + 1, int(im.shape[1] / stride) + 1))

im_pad = np.pad(im, 1, "constant")

for i in range(0, im.shape[0], stride):
    for j in range(0, im.shape[1], stride):
        try:
            new_image_stride3_vert[int(i / stride), int(j / stride)] = (
                im_pad[i - 1, j - 1] * filter2[0, 0]
                + im_pad[i - 1, j] * filter2[0, 1]
                + im_pad[i - 1, j + 1] * filter2[0, 2]
                + im_pad[i, j - 1] * filter2[1, 0]
                + im_pad[i, j] * filter2[1, 1]
                + im_pad[i, j + 1] * filter2[1, 2]
                + im_pad[i + 1, j - 1] * filter2[2, 0]
                + im_pad[i + 1, j] * filter2[2, 1]
                + im_pad[i + 1, j + 1] * filter2[2, 2]
            )
        except:
            pass

plt.axis("off")
plt.imshow(new_image_stride3_vert, cmap="Greys")
print(f"Output shape with stride=3: {new_image_stride3_vert.shape}")
```

## Comparison: Stride 2 vs Stride 3

```python
# Read the stride=2 results from earlier cells
filter1 = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
stride = 2
new_image_stride2 = np.zeros((int(im.shape[0] / stride), int(im.shape[1] / stride)))
im_pad = np.pad(im, 1, "constant")

for i in range(0, im.shape[0], stride):
    for j in range(0, im.shape[1], stride):
        try:
            new_image_stride2[int(i / stride), int(j / stride)] = (
                im_pad[i - 1, j - 1] * filter1[0, 0]
                + im_pad[i - 1, j] * filter1[0, 1]
                + im_pad[i - 1, j + 1] * filter1[0, 2]
                + im_pad[i, j - 1] * filter1[1, 0]
                + im_pad[i, j] * filter1[1, 1]
                + im_pad[i, j + 1] * filter1[1, 2]
                + im_pad[i + 1, j - 1] * filter1[2, 0]
                + im_pad[i + 1, j] * filter1[2, 1]
                + im_pad[i + 1, j + 1] * filter1[2, 2]
            )
        except:
            pass

# Compare stride=2 vs stride=3 side by side
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(im, cmap='gray')
axes[0].set_title('Original Image\n(64Ã—64)', fontweight='bold', fontsize=12)
axes[0].axis('off')

axes[1].imshow(new_image_stride2, cmap='Greys')
axes[1].set_title(f'Horizontal Edges - Stride=2 (Original)\n({new_image_stride2.shape[0]}Ã—{new_image_stride2.shape[1]})', 
                  fontweight='bold', fontsize=12)
axes[1].axis('off')

axes[2].imshow(new_image_stride3, cmap='Greys')
axes[2].set_title(f'Horizontal Edges - Stride=3 (Custom)\n({new_image_stride3.shape[0]}Ã—{new_image_stride3.shape[1]})', 
                  fontweight='bold', fontsize=12)
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

## Summary: Comparing Stride=2 vs Stride=3

### What I Observed in the Coffee Image

**Stride=2 (Original):**  
When the filter moved 2 pixels at a time, our 64Ã—64 coffee image became a 32Ã—32 image that's **75% smaller**. Looking at the results, the edges still look **sharp and clear**. I could easily see the coffee cup and plate outlines, which means we kept most of the important details even though we made the image much smaller.

**Stride=3 (Custom):**  
When I increased it to stride=3 (jumping 3 pixels each time), the image got even smaller just 22Ã—22 pixels. That's **88% smaller** than the original. The edges look **chunkier and more pixelated** now. The main shapes are still there, but you can tell we lost some of the finer details because we're skipping over more information.

### Key Differences

**Size Comparison:**

| Configuration | Output Size | Size Reduction | Visual Quality |
|--------------|-------------|----------------|----------------|
| Stride=2 | 32Ã—32 (1,024 pixels) | 75% smaller | Sharp, clear edges |
| Stride=3 | 22Ã—22 (484 pixels) | 88% smaller | Blockier, pixelated |

**What This Means:**
- **Stride=2** keeps the edges looking good while making the image 4 times smaller in terms of data to process
- **Stride=3** makes things even smaller (9 times less data) but the quality drops noticeably
- The difference between them: Stride=3 has about **53% fewer pixels** than stride=2's output

### My Conclusion

For edge detection on a 64Ã—64 image like our coffee cup, **stride=2 is the better choice**. It gives us a good balance. The image is much smaller and faster to work with, but we don't lose the important edge details. 

Stride=3 might be okay for really big images where there's lots of extra information to spare, but for our small coffee image, it cuts out too much detail. The edges start looking blocky and we miss some of the finer features that make the image recognizable.

The key lesson: **bigger stride = faster processing, but also = more information loss**. We have to find the right balance for your specific image size and what we are trying to detect.
